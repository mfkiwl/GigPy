```python
from gigpy import _ExtractData as ed
from gigpy import _utils as ut
import os

def extract_sar_delay_unr(data0):
    date, stations, imaging_time, data_path, out_path = data0
    unrSec = str(round(int(imaging_time)/300)*300)
    
    if len(unrSec) ==1: unrSec = '0000' + unrSec
    elif len(unrSec) ==3: unrSec = '00' + unrSec
    elif len(unrSec) ==4: unrSec = '0' + unrSec
        
    year, doy = ut.yyyymmdd2yyyyddd(date)
    extract_time = year[2:4] + ':' + doy + ':' + unrSec
    
    out_file = out_path + '/SAR_GPS_Trop_' + date
    if os.path.isfile(out_file):
        os.remove(out_file)
    nn = len(stations)
    for i in range(nn):
        file0 = data_path + '/' + stations[i].upper() + doy + '0.' + year[2:4] + 'zpd.gz'
        #print(file0)
        STR0 = '"' + stations[i].upper() + ' ' + extract_time + '"'
        print(STR0)
        call_str = 'zgrep -i ' + STR0 + ' ' + file0 + ' >>' + out_file
        print(call_str)
        #call_str = 'zgrep -i ' + STR0 + ' ' + file0
        os.system(call_str)
    
    return

gps_file = '/Users/caoy0a/Documents/SCRATCH/LosAngelesT71F479D/gps_station_info.txt'
A = ut.read_txt2array(gps_file)
stations = A[:,0]
data_path = '/Users/caoy0a/Documents/SCRATCH/LosAngelesT71F479D/gigpy/atm/raw/20150101'
out_path = '/Users/caoy0a/Documents/SCRATCH/LosAngelesT71F479D/gigpy/atm/sar_raw'

data0 = ['20150101', stations, 49942, data_path, out_path]
extract_sar_delay_unr(data0)
```
